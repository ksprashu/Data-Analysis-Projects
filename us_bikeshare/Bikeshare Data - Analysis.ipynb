```python
import pandas as pd
import numpy as np

cities = ['chicago', 'washington', 'new_york']
files = ['chicago.csv', 'washington.csv', 'new_york_city.csv']
data = []

# load all the files
for city, file in zip(cities, files):
    data.append((city, pd.read_csv(file)))

# chicago = pd.read_csv('chicago.csv')
# washington = pd.read_csv('washington.csv')
# new_york = pd.read_csv('new_york_city.csv')

print('done')
print(data)
```

```python
# can I check that some data sets don't have gender?

for city, df in data:
    print('Checking: {} has Gender column : {}'.format(city, 'Gender' in df.columns))
```

```python
# checking that some data sets don't have birth year

for city, df in data:
    print('Checking: {} has Birth Year column : {}'.format(city, 'Birth Year' in df.columns))
```

```python
# getting unique counts and other data for birth year

for city, df in data:
    if 'Birth Year' in df.columns:
        count = df['Birth Year'].value_counts()
        
        print('Birth Year data for {}'.format(city))
        print(count)
        
        print('\nMax birth year for {} is {}'.format(city, df['Birth Year'].max()))
        print('Min birth year for {} is {}'.format(city, df['Birth Year'].min()))
        print('Most repeated birth year for {} is {}'.format(city, count.index[0]))
        
    else:
        print('\n{} has no Birth Year\n'.format(city))
        
```

```python
# check that all cities have only month data from Jan to Jun

for city, df in data:
    # clean up start time and end time data
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['Start Time'].head()
    df['Month'] = df['Start Time'].dt.month_name()
    df['Day'] = df['Start Time'].dt.day_name()
    
    df['End Time'] = pd.to_datetime(df['End Time'])
    df['End Time'].head()
    
    print('Unique Months in {} are {}'.format(city, df['Month'].unique()))
    print('Unique Days in {} are {}\n'.format(city, df['Day'].unique()))
```

```python
# Birth year seems to have NaN in the data??

for city, df in data:
    if 'Birth Year' in df.columns:
        print('total count for {} is {}'.format(city, df['Birth Year'].count()))
        print('NaN count for {} is {}'.format(city, df['Birth Year'].isnull().sum()))
        
```

```python
# try to clean up birth year data from float to int

for city, df in data:
    if 'Birth Year' in df.columns:
        print('original data for {} is '.format(city))
        print(df['Birth Year'].head())

        df['Birth Year'] = pd.to_numeric(df['Birth Year'], downcast='integer', errors='ignore')
        print('cleaned up data for {} is'.format(city))
        print(df['Birth Year'].head())
```

```python
# getting a combination of data

for city, df in data:
    print('Data for {}\n'.format(city))
    print(df['Start Station'].value_counts().head())
    print('')
    print(df['End Station'].value_counts().head())
    
    print('\nGrouped sizes')
    groupies = df.groupby(['Start Station', 'End Station'], as_index=False).size()
    max_row = groupies.idxmax()
    max_size = groupies.max()
    
    print('The grouped max station pair is {} with occurance of {} times'.format(max_row, max_size))
    
    print('\nNow calculating by concatenating the columns')
    calc = pd.Series(df['Start Station'] + '_' + df['End Station'])
    print(calc.head())
    
    
```
