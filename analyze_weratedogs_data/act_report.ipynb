# Analysis of Dog Ratings by WeRateDogs

## Background
WeRateDogs is a famous twitter account that ratings various dogs based on the images that are submitted by other users. This is a highly followed account with over 8.3M followers. WeRateDogs usually provides a nice commentary and funny description about the dogs themselves while adding a rating. 

Ratings are on a scale of 10, and many dogs get a rating of 11/10 and upwards. This is one of the factors of this account that followers love.

## Objective
The objective of this analysis is to understand and analyse the tweets from this user. It would be interesting to see the kinds of ratings that are given and the distribtion of the ratings themselves across the dataset. It would also be interesting to see the breeds of dogs that are being shared and rated and understand what are the most popular breeds of dogs on twitter and which are the highest rated breeds. 

## Preparation
An archive of the latest 5000 tweets of this user has been obtained by Udacity. The archive of tweets was further put through a image recognition neural network by Udacity which predicted whether the tweets were about dogs or other animals (based on the images associated with each tweet). 

Subsequently, I then pulled together the actual tweets corresponding to each line item of the datasets from twitter using their API. This data that was gathered was Assessed and Cleaned over multiple iteration to end up with a clean and high quality dataset that we can now do analysis on top of.

### Spread of ratings
It is seen that the highest ratings awarded to dogs are 14. Most of the ratings fall between 10 and 12, while some dogs get a rating of 13, and even fewer dogs get a rating of 14. Of course in the original raw data, it was seen that some ratings are really high such as 666 and 1776. However these are mostly satire and can be ignored.

![Histogram of ratings](hist_rating.png)

### Correlation between favourite & retweet counts
It is interesting to see whether the tweets that get a high amount of retweets are also highly favourited, and vice versa. Upon analysis of the data, indeed there is a very strong correlation betweent the number of retweets and the number of favourites. 

![Correlation between favourites and retweets](corr_ret_fav.png)

### Relation between ratings and favourite / retweet counts
Now that we see that favourite count and retweet counts are pretty much the same in representing the popularity of the tweet, it would be interesting to see whether the ratings themselves are in proportion i.e. do the tweets with highest ratings get shared around or upvoted a lot? This would also indicate in a way the critics ratings in comparision to the user ratings (as a measure of social promotion).

![Correlation between ratings and social upvotes](rel_rating_fav.png)

It does look like mostly the ratings of 13 have got the largest number of retweets and/or favourites, as compared the rating of 14. Of course, between 10 and 13, the number of social upvotes is directly proportional to the rating itself.
